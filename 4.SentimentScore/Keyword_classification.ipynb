```python
import pandas as pd
import numpy as np
```

```python
test_data_상의 = pd.read_csv('Example_review/test_data.csv')
test_data_상의 = test_data_상의[['Review']]
test_data_상의
```

```python
test_data_상의['Review'] = test_data_상의['Review'].str.replace("[^ㄱ-ㅎㅏ-ㅣ가-힣 ]","")
print(test_data_상의.isnull().sum())
```

```python
import warnings
warnings.filterwarnings(action='ignore')

test_data_상의['색감'] = '0'
test_data_상의['핏'] = '0'
test_data_상의['디자인'] = '0'
test_data_상의['가격'] = '0'
test_data_상의['재질'] = '0'
test_data_상의['배송'] = '0'
```

```python
l = []

for i in range(0, len(test_data_상의['Review'])):
    a = test_data_상의['Review'][i]
    l.append(a)
    
print(len(l))
```

```python
색감_result = test_data_상의.query('Review.str.contains("색감|색상|오트밀|그레이|색|색도|색깔|블랙|회색|블루|아이보리|흰색|컬러|채도|명도|어두운|파란색|연한|화이트|멜란지|검정|연한")', engine='python')
색감_list = list(set(list(색감_result['Review'])))
for i in range(0, len(test_data_상의['Review'])):
    if l[i] in 색감_list:
        test_data_상의['색감'][i] = test_data_상의['색감'][i].replace('0', '1')
        
print(test_data_상의['색감'].value_counts())


핏_result = test_data_상의.query('Review.str.contains("사이즈|핏|기장|길이|폼|소매|핏감|루즈|루즈핏|널널|스펙|정사|이즈|라지|활용|체형|키|정핏|실루엣|와이드|흐물흐물|몸무게|분위기|넥라인|넥부분")', engine='python')
핏_list = list(set(list(핏_result['Review'])))
for i in range(0, len(test_data_상의['Review'])):
    if l[i] in 핏_list:
        test_data_상의['핏'][i] = test_data_상의['핏'][i].replace('0', '1')
        
print(test_data_상의['핏'].value_counts())


디자인_result = test_data_상의.query('Review.str.contains("디자인|스타일|심플|트렌드|트랜디|연출|코디|스탠다드|유행|꾸안꾸룩|데님|네추럴|트렌디|무드|휘뚜르|마뚜르|매치|매칭")', engine='python')
디자인_list = list(set(list(디자인_result['Review'])))
for i in range(0, len(test_data_상의['Review'])):
    if l[i] in 디자인_list:
        test_data_상의['디자인'][i] = test_data_상의['디자인'][i].replace('0', '1')
        
print(test_data_상의['디자인'].value_counts())


가격_result = test_data_상의.query('Review.str.contains("가격|가성|할인|가성비|블프|블랙프라이데이")', engine='python')
가격_list = list(set(list(가격_result['Review'])))
for i in range(0, len(test_data_상의['Review'])):
    if l[i] in 가격_list:
        test_data_상의['가격'][i] = test_data_상의['가격'][i].replace('0', '1')
        
print(test_data_상의['가격'].value_counts())


재질_result = test_data_상의.query('Review.str.contains("재질|두께|촉감|따뜻|퀄리티|보풀|소재|원단|니트|집업|겨울|가을|지퍼|품질|마감|실밥|밑단|옷감|까슬까슬|부드러움|관리|세탁|줄어듬|먼지|쫀쫀|보온|무게|한겨울|냄새|건조기|짜임새|기모|감촉|싸구려")', engine='python')
재질_list = list(set(list(재질_result['Review'])))
for i in range(0, len(test_data_상의['Review'])):
    if l[i] in 재질_list:
        test_data_상의['재질'][i] = test_data_상의['재질'][i].replace('0', '1')
        
print(test_data_상의['재질'].value_counts())


배송_result = test_data_상의.query('Review.str.contains("배송|포장|교환|딜레이|검수|불량|불량품|서비스|주문|반품")', engine='python')
배송_list = list(set(list(배송_result['Review'])))
for i in range(0, len(test_data_상의['Review'])):
    if l[i] in 배송_list:
        test_data_상의['배송'][i] =test_data_상의['배송'][i].replace('0', '1')
        
print(test_data_상의['배송'].value_counts())
```
